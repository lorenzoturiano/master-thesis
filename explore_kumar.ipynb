```python
import numpy as np
import scanpy as sc
```

```python
# Load the .h5ad file
adata = sc.read_h5ad("datasets/single-cell/7b34508e-c302-489a-bba8-9e963e350daf.h5ad")
```

```python
n_cells, n_genes = adata.shape
print(f"Number of cells: {n_cells}\nNumber of genes: {n_genes}")
```

```python
categories = list(set(adata.obs["cell_type"].to_list()))
n_cat = len(categories)

print(f"There are {n_cat} categories")
```

```python
categories_count = {type: 0 for type in categories}

for sample_type in adata.obs["cell_type"]:
    categories_count[sample_type] += 1

for type in categories:
    print(f"{type}: {categories_count[type]}")
```

```python
min_percentage = 0.01
min_count = int(np.ceil(n_cells * min_percentage))

for k,v in categories_count.items():
    if v < min_count:
        print(f"{k} cell-type has less than {min_count} counts ({v})")
```
