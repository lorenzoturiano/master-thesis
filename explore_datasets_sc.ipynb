```python
import scanpy as sc
```

```python
# Load the .h5ad file
adata_kidney_2= sc.read_h5ad("datasets/single-cell/9beaa89c-608a-4ef7-a816-2e6d42b1a0c2.h5ad")
```

```python
adata_kidney_2_normal = adata_kidney_2[adata_kidney_2.obs["disease"] == "normal"].copy()
df_kidney_2_normal = adata_kidney_2_normal.obs
```

```python
df_kidney_2_normal.shape
```

```python
cell_types = list(set(df_kidney_2_normal["cell_type"].to_list()))

classes = list(set(df_kidney_2_normal["Class"].to_list()))

cell_types, classes
```

```python
# change this if your labels are in a different column (e.g., "celltype_l2")
col = "cell_type"

counts_dict = (
    adata_kidney_2_normal.obs[col]
    .value_counts()
    .reindex(cell_types, fill_value=0)
    .astype(int)
    .to_dict()
)

counts_dict
```

```python
# change this if your labels are in a different column (e.g., "celltype_l2")
col = "Class"

counts_dict = (
    adata_kidney_2_normal.obs[col]
    .value_counts()
    .reindex(classes, fill_value=0)
    .astype(int)
    .to_dict()
)

counts_dict
```
